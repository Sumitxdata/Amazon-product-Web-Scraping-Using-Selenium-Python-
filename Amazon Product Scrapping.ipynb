```python
# Import Libreries
from selenium import webdriver
import pandas as pd
import numpy as np
import urllib
```

```python
# Launch the Browser Engine
bulk = webdriver.Chrome()
```

```python
# Open The source url
bulk.get('https://www.amazon.eg/s?k=samsung&i=aps&ref=nb_sb_ss_ts-doa-p_1_3&crid=2VD3J7AABKQJP&sprefix=sam%2Caps%2C300')
```

```python
# get the product details as every property is alone 
bulk_images = bulk.find_elements('xpath',"//img[@class='s-image']")
bulk_description = bulk.find_elements('xpath',"//span[@class='a-size-base-plus a-color-base a-text-normal']")
bulk_prices = bulk.find_elements('xpath',"//span[@class='a-price-whole']")
bulk_rate = bulk.find_elements('xpath',"//div[@class='a-row a-size-small']")
```

```python
# Product Images
for i in bulk_images:
    print(i.get_attribute('src'))
```

```python
# Product description
for i in bulk_description:
    print(i.text)
```

```python
# Product Price
for i in bulk_prices:
    print(i.text)
```

```python
# Product Rate and Rate people
rates = []
num_raters = []
for rate_line in bulk_rate:
    star_rating = rate_line.get_attribute("textContent")
    if "out of 5 stars" in star_rating:
        print(star_rating.split()[0])
        rates.append(star_rating.split()[0])
        num_raters.append(star_rating.split()[-1])
```

```python
print(num_raters)
```

```python
# The length of every item in the product category 
print(len(bulk_images))
print(len(bulk_description))
print(len(bulk_prices))
print(len(rates))
print(len(num_raters))
```

#### So we have to implement try and except handling to prevernt and handle missing values

```python
products_Sections = bulk.find_elements('xpath',"//div[@class='a-section a-spacing-base']")
for section in products_Sections:
    try:
        title=section.find_element('xpath',".//span[@class='a-size-base-plus a-color-base a-text-normal']")
        print(title.text)
        price=section.find_element('xpath',".//span[@class='a-price-whole']")
        print(price.text)
        image_src=section.find_element('xpath',".//img[@class='s-image']")
        url=image_src.get_attribute('src')
        print(url)
        rating=section.find_element('xpath',".//div[@class='a-row a-size-small']")
        
        
        star_rating = rating.get_attribute("textContent")
        if "out of 5 stars" in star_rating:
            print(star_rating.split()[0])

        total_rating=section.find_element('xpath',".//span[@class='a-size-base s-underline-text']")
        print(total_rating.text)
#         urllib.request.urlretrieve(url,"{}.jpg".format(title.text))
    except:
        print('error')
```

```python
products_Sections = bulk.find_elements('xpath',"//div[@class='a-section a-spacing-base']")
title_list = []
price_list = []
image_list = []
rate_list = []
raters_list = []
for section in products_Sections:
    
    try:
        title=section.find_element('xpath',".//span[@class='a-size-base-plus a-color-base a-text-normal']")
        print(title.text.strip())
        title_list.append(title.text.strip())
    except:
        title_list.append(np.NAN)

    try:
        price=section.find_element('xpath',".//span[@class='a-price-whole']")
        print(price.text.strip())
        price_list.append(price.text.strip())
        
    except:
        price_list.append(np.NAN)
        
    try:
        image_src=section.find_element('xpath',".//img[@class='s-image']")
        url=image_src.get_attribute('src')
        print(url)
        image_list.append(url)
        
    except:
        image_list.append(np.NAN)
        
    try:      
        rating=section.find_element('xpath',".//div[@class='a-row a-size-small']")
        star_rating = rating.get_attribute("textContent")
        if "out of 5 stars" in star_rating:
            print(star_rating.split()[0])
            rate_list.append(star_rating.split()[0])
 
    except:
        rate_list.append(np.NAN)
    
    try:
        total_rating=section.find_element('xpath',".//span[@class='a-size-base s-underline-text']")
        print(total_rating.text.strip())
        raters_list.append(total_rating.text.strip())
    except:
        raters_list.append(np.NAN)
    
# the below line for extracting and downloading all product images
    # urllib.request.urlretrieve(url,"{}.jpg".format(title.text))    
```

```python
# So we have right now equevilant lists
print(len(title_list))
print(len(price_list))
print(len(image_list))
print(len(rate_list))
print(len(raters_list))
```

```python
data = {'Description': title_list, 'Price': price_list, 'Image URL': image_list, 'Rate': rate_list, 'Total Rate': raters_list}
df = pd.DataFrame(data)
```
